### 3.2.1. <a id='toc3_2_1_'></a>[Feature selection](#toc0_)

```python
# Feature selection
all_features_class = ['id', 'customer_type', 'age', 'class', 'flight_distance',
       'inflight_wifi_service', 'departure_arrival_time_convenient',
       'ease_of_online_booking', 'gate_location', 'food_and_drink',
       'online_boarding', 'seat_comfort', 'inflight_entertainment',
       'on_board_service', 'leg_room_service', 'baggage_handling',
       'checkin_service', 'inflight_service', 'cleanliness',
       'departure_delay_in_minutes', 'arrival_delay_in_minutes',
       'gender_Female', 'gender_Male', 'type_of_travel_business_travel',
       'type_of_travel_personal_travel']

selected_features_class = ['customer_type', 'age', 'class', 'flight_distance',
       'inflight_wifi_service', 'departure_arrival_time_convenient',
       'ease_of_online_booking', 'gate_location', 'food_and_drink',
       'online_boarding', 'seat_comfort', 'inflight_entertainment',
       'on_board_service', 'leg_room_service', 'baggage_handling',
       'checkin_service', 'inflight_service', 'departure_delay_in_minutes', 
       'arrival_delay_in_minutes', 'gender_Female', 'gender_Male', 
       'type_of_travel_personal_travel']
```

```python
X_train_cla = X_train_cla_raw[selected_features_class]
X_val_cla = X_val_cla_raw[selected_features_class]
X_test_cla = X_test_cla_raw[selected_features_class]
```

### 3.2.2. <a id='toc3_2_2_'></a>[K-Nearest Neighbors](#toc0_)

#### 3.2.2.1. <a id='toc3_2_2_1_'></a>[Training dataset](#toc0_)

```python
X_train_cla.head(1)
```

```python
k_neighbors = 3

# Model calling
knn = nb.KNeighborsClassifier(n_neighbors=k_neighbors)

# Model Fit
knn.fit(np.array(X_train_cla), np.array(y_train_cla))

# Prediction
ypred_train_knn = knn.predict(X_train_cla)
```

```python
# Exporting KNN model
# knn_filename = 'knn_model.sav'
# pickle.dump(knn, open(knn_filename, 'wb'))
```

```python
# Metrics
knn_train_precision = mt.precision_score(y_train_cla, ypred_train_knn)
knn_train_accuracy = mt.accuracy_score(y_train_cla, ypred_train_knn)
knn_train_recall = mt.recall_score(y_train_cla, ypred_train_knn)
knn_train_F1 = mt.f1_score(y_train_cla, ypred_train_knn)

print('KNN - Training Dataset')
print(f'Accuracy: {round(knn_train_accuracy, 3)}')
print(f'Precision: {round(knn_train_precision, 3)}')
print(f'Recall: {round(knn_train_recall, 3)}')
print(f'F1-Score: {round(knn_train_F1, 3)}')
```

#### 3.2.2.2. <a id='toc3_2_2_2_'></a>[Validation dataset](#toc0_)

```python
# Predict
ypred_val_knn = knn.predict(X_val_cla)
```

```python
# Metrics
knn_val_precision = mt.precision_score(y_val_cla, ypred_val_knn)
knn_val_accuracy = mt.accuracy_score(y_val_cla, ypred_val_knn)
knn_val_recall = mt.recall_score(y_val_cla, ypred_val_knn)
knn_val_F1 = mt.f1_score(y_val_cla, ypred_val_knn)

print('KNN - Validation Dataset')
print(f'Accuracy: {round(knn_val_accuracy, 3)}')
print(f'Precision: {round(knn_val_precision, 3)}')
print(f'Recall: {round(knn_val_recall, 3)}')
print(f'F1-Score: {round(knn_val_F1, 3)}')
```

#### 3.2.2.3. <a id='toc3_2_2_3_'></a>[Test dataset](#toc0_)

```python
# Predict
ypred_test_knn = knn.predict(X_test_cla)
```

```python
# Metrics
knn_test_precision = mt.precision_score(y_test_cla, ypred_test_knn)
knn_test_accuracy = mt.accuracy_score(y_test_cla, ypred_test_knn)
knn_test_recall = mt.recall_score(y_test_cla, ypred_test_knn)
knn_test_F1 = mt.f1_score(y_test_cla, ypred_test_knn)

print('KNN - Test Dataset')
print(f'Accuracy: {round(knn_test_accuracy, 3)}')
print(f'Precision: {round(knn_test_precision, 3)}')
print(f'Recall: {round(knn_test_recall, 3)}')
print(f'F1-Score: {round(knn_test_F1, 3)}')
```
